##Q

```python
# set the histogram, mean and median
sns.displot(df["Q"], kde=True)
plt.axvline(x=df.Q.mean(), linewidth=3, color='g', label="mean", alpha=0.5)
plt.axvline(x=df.Q.median(), linewidth=3, color='y', label="median", alpha=0.5)

# set title, legends and labels
plt.xlabel("Lưu lượng nước")
plt.ylabel("Số lượng")
plt.title("Biểu đồ phân phối của lưu lượng nước", size=14)
plt.legend(['KDE',"Mean", "Median"]);

print(f' Mean Q value: {df.Q.mean()} \n Median Q value: {df.Q.median()} \n Min Q value: {df.Q.min()} \n Max Q value: {df.Q.max()}')
```

Nhận xét: Lưu lượng nước trung bình năm 2019 là 75,1. Lượng nước cao nhất trong năm rơi vào khoảng 600. Tần số lưu lượng nước rơi vào 20 đến 50 là nhiều nhất, lưu lượng nước càng cao có xu hướng xuất hiện ít hơn. Chuỗi có xu hướng phân phối về phía bên trái vì median nhỏ mean.

```python
df_infl = df3
for col in df_infl.columns:
  if df_infl[col].dtypes == np.int64 or df_infl[col].dtypes == np.float64: 
    continue
  else:
    try:
        df_infl = df_infl.drop(labels=col, axis=1)
    except Exception:
        continue
```

```python
df_infl
```

```python
corrMatrix_c5 = df_infl.corr()
corrMatrix_c5
```

```python
correlation1 = df_infl.corr(method='pearson')
cfig1 = plt.subplots(figsize=(5,5))

mask = np.zeros_like(correlation1)
mask[np.triu_indices_from(mask)] = True

# Draw the heatmap with the mask
#sns.heatmap(corr_matrix, mask=mask, square=True)

sns.heatmap(correlation1,vmax=1,square=True,annot=True,cmap='YlGnBu', mask=mask)
```
