# PLOT ĐỘ DAO ĐỘNG MỰC NƯỚC TRÊN BỘ DỮ LIỆU HOÀN CHỈNH




```python
# chọn ngày giờ theo left và right
def followhour(df_hour, l, r):
  df_ = df_hour[(df_hour['Timestamp']>=l) & (df_hour['Timestamp']<=r)]
  return df_ 
```

```python
df_hour
```

```python
df_day1 = followhour(df_hour, '2019-12-29 00:00:00', '2019-12-29 23:00:00')
df_day1
```

```python
# Chu kỳ của mực nước theo giờ trong ngày
# chọn 3 ngày cuối năm 2019 29/12/2019, ngày 30/12/2019, 31/12/2019 

#df_day1 = followhour(df_hour, '2019-12-29 0:00:00', '2019-12-29 23:00:00')
#df_day2 = followhour(df_hour, '2019-12-30 0:00:00', '2019-12-30 23:00:00')
#df_day3 = followhour(df_hour, '2019-12-31 0:00:00', '2019-12-31 23:00:00')

df_day1 = followhour(df_hour, '2019-12-29 0:00:00', '2019-12-29 23:00:00')
df_day2 = followhour(df_hour, '2019-12-30 0:00:00', '2019-12-30 23:00:00')
df_day3 = followhour(df_hour, '2019-12-31 0:00:00', '2019-12-31 23:00:00')


fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(40, 26))
fig.suptitle('Chu kỳ của mực nước theo giờ trong ngày', fontsize=30)

ax1.plot(df_day1['Timestamp'], df_day1['Level'], color='blue')
ax1.set_title('29/12/2019', color='red')

ax2.plot(df_day2['Timestamp'], df_day2['Level'], color='blue')
ax2.set_title('30/12/2019', color='red')

ax3.plot(df_day3['Timestamp'], df_day3['Level'], color='blue')
ax3.set_title('31/12/2019', color='red')

plt.rcParams["font.size"] = "16"
plt.show()
```

```python
# Chu kỳ của mực nước theo giờ trong tháng
df_month = df_day[['Date', 'Level', 'Month']]
df_month = df_month.groupby('Month', as_index=False).mean()
```

# PLOT ĐỘ DAO ĐỘNG MỰC NƯỚC TRÊN BỘ DỮ LIỆU THÔ

```python
# dùng bộ dữ liệu theo ngày, thêm các cột day month year
df_day_['Day'] = pd.to_datetime(df_day_['Day'])

#Create 3 additional columns
df_day_['day'] = df_day_['Day'].dt.day
df_day_['month'] = df_day_['Day'].dt.month
df_day_['year'] = df_day_['Day'].dt.year
```

```python
df_day_
```

```python
# Plot mực nước
plt.subplots(figsize=(40, 10))
sns.lineplot(data=df_day_, x='Day', y='Level', hue='year')
```

* Mực nước trung bình nửa đầu của từng năm đều cao

# PLOT ĐỘ DAO ĐỘNG CỦA CÁC CHỈ SỐ

```python
# chọn ngày 31/12/2019
day3112 = followhour(df_hour, '2019-12-31 00:00:00', '2019-12-31 23:00:00')	
```

```python
day3112
```

```python
# scale giá trị
day3112['Q']=day3112['Q']/day3112['Q'].max()
day3112['Conductivity']=day3112['Conductivity']/day3112['Conductivity'].max()
day3112['NO3']=day3112['NO3']/day3112['NO3'].max()
day3112['Temp']=day3112['Temp']/day3112['Temp'].max()
day3112['Turbidity']=day3112['Turbidity']/day3112['Turbidity'].max()
day3112['Level']=day3112['Level']/day3112['Level'].max()
```

```python
day3112
```

```python
day3112.plot(x='Timestamp', y=['Conductivity',	'NO3', 'Temp', 'Turbidity', 'Level'], kind="line", figsize=(30, 8))
plt.ylabel('Other Feature')
plt.xlabel('Time')
plt.title('Ngày 31/12/2019', color='brown')
plt.show()
```
