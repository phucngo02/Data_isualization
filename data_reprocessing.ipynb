## Data theo ngay

```python
ngay = df.copy()
ngay = ngay[(ngay['Timestamp']>='2019-01-01T00:00:00') & (ngay['Timestamp']<='2019-12-31T23:00:00')]
ngay_temp = ngay.copy()
ngay = pd.DataFrame(ngay['Timestamp'])
ngay = ngay.join(ngay['Timestamp'].str.split('T', 1, expand=True).rename(columns={0:'Date', 1:'Time'}))
ngay
```

```python
ngay.drop('Timestamp', axis=1, inplace=True)
ngay[['Q',	'Conductivity',	'NO3',	'Temp',	'Turbidity',	'Level',	'Dayofweek',	'Month']] = ngay_temp[['Q',	'Conductivity',	'NO3',	'Temp',	'Turbidity',	'Level',	'Dayofweek',	'Month']]
ngay
```

```python
ngay_groupby = ngay.groupby('Date').mean()
ngay_groupby
```

```python
ngay_groupby.info()
```

```python
ngay_groupby.to_csv('Mulgrave_2019_Theo_Ngay_NoMiss.csv')
```
